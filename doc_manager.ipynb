```python
import os
import psutil
import pathlib
import pandas as pd
import numpy as np
```

```python
def file_seperator():
    try:
        user_input = input("Please enter your directory or file: ").strip()
        normalized = user_input.replace("\\", "/")
        if len(normalized) >= 2 and normalized[1] == ":" and normalized[0].isalpha():
            normalized = f"/mnt/{normalized[0].lower()}{normalized[2:]}"
        file_path = pathlib.Path(normalized).expanduser()
        if not file_path.is_absolute():
            file_path = (pathlib.Path.cwd() / file_path).resolve()
        print(file_path)
        if file_path.is_dir():
            for item in file_path.iterdir():
                print(item.stat())
        elif file_path.is_file():
            print(file_path.stat())
        else:
            raise ValueError(f"File or directory does not exist: {file_path}")
    except Exception as e:
        print(f"Error with script: {e}")
```
