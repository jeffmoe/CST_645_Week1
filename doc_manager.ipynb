```python
import os
import psutil
import pathlib
import numpy as np
```

```python
def extract_file_details():
    """
    Extracts the base filename and extension from a given file path or filename.

    Parameters:
    filename (str): The file path or filename provided by the user.

    Returns:
    tuple: A tuple containing:
        - name (str): The filename without extension
        - extension (str): The file extension (empty string if none exists)
    """
    user_input = input("Please provide the path or filename:") #getting input from the user
    try: # try loop so we can output errors if they happen
        file_name = os.path.basename(user_input)
        if file_name.startswith('.') and file_name.count('.') == 1: #check for hidden files
            return file_name, ''
        if '.' in file_name: #checking for period separators which are the standard format for file extensions
            name = file_name.split('.')
            name, extension = name[0], name[1:]
            return name, extension
        else:
            return file_name, '' #if there are no periods then it just returns the inputted value as a string and a blank string
    except Exception as e:
        print(f"Error getting file details: {e}") #error handler

```

```python
extract_file_details()
```
